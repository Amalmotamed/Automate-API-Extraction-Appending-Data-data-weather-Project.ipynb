```python
#extract data from rest country api
import pandas as pd
import requests

url = "https://restcountries.com/v3.1/all?fields=name,languages,capital,region,subregion,translation"
response = requests.get(url)
data = response.json()
countries = []
for country in data:
    name=country["name"]["common"]

    if "languages" in country:
        languages=list(country["languages"])

    if "translation" in country:
        translation =list( country["translation"]["name"]["common"])
    else :
        translation= None

    if "region" in country:
        region=list(country["region"])

    if "subregion" in country:
        subregion=list(country["subregion"])

    if "capital" in country:
        capital =list(country["capital"])
    
    countries.append({"country": name ,"capital":capital,"languages": languages,"region": region,"subregion": subregion ,"translation": translation})

df = pd.DataFrame(countries)


```
